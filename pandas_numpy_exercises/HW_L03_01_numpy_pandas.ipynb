# 2. Data Preprocessing and Numerical Operation

### Create Dataset

```python
np.random.seed(101)
scores = np.random.randint(50, 100, size=(15, 3))
scores
```

```python
scores[1][0]
```

### Data Analize

```python
col_mean = np.mean(scores, axis=0)
col_mean
```

```python
std_scores = np.std(scores, axis=0)
std_scores
```

```python
max_score = np.max(scores, axis=0)[2]
max_score
```

```python
row_mean = np.mean(scores, axis=1)
row_mean
```

### Broadcasting

```python
curved_scores = np.array(5)
print(curved_scores + scores[:, 5])
scores
```

### Normalization

```python
normalized_scores = (scores - col_mean) / std_scores
normalized_scores
```

```python
min_normalized_scores = np.mean(normalized_scores, axis=0)
print(min_normalized_scores)
std_normalized_scores = np.std(normalized_scores, axis=0)
print(std_normalized_scores)
```

# 3. Data Filtering & Feature Engineering 

### Boolean Mask

```python
high_scores = scores > 95
high_scores
```

```python
high_scores_quantity = np.sum(scores > 95)
high_scores_quantity
```

```python
high_performance_score = scores[scores>95]
high_performance_score
```

```python
low_scores = scores[scores[:, 0] <= 70]
mean_low = np.mean(low_scores[:, 1])
print("students who get score equal or under 70 at exam 1:\n", low_scores)
print("avarage scores of these students:\n", mean_low)
```

### Feature Engineering

```python
ave_scores = np.mean(scores, axis=1)
print(ave_scores)
pass_fail = np.bool(np.where(ave_scores<65, False, ave_scores))
pass_fail
```

### Complicated Filtering

```python
top_students = np.all(scores>=60, axis=1)
if not np.any(top_students):
    print("No top student!")
else:
    print(np.where(top_students))
    print(scores[top_students])
```

# 5. Cleaning and PreProcessing

### Identifing Missing Values

```python
df.isnull().sum()
```

```python
df_cleaned = df.drop('Cabin', axis='columns')
df_cleaned
```

### Handling Missing Values

```python
median = df['Age'].median()
median
```

```python
df_cleaned['Age'] = df_cleaned['Age'].fillna(median)
df_cleaned
```

```python
mode = df['Embarked'].mode()
mode
```

```python
df_cleaned['Embarked'] = df_cleaned['Embarked'].fillna(mode[0])
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

### Process Finalization

```python
df_cleaned['Pclass'] = df_cleaned['Pclass'].astype({'Pclass':str})
df_cleaned['Pclass']
```

# 6. Feature Engineering and Group Analysis 

### Creating New Features from Existing Ones

```python
df_cleaned['FamilySize'] = df_cleaned['SibSp'] + df_cleaned['Parch']
df_cleaned
```

### Extracting Information to Build Features

```python
df_cleaned['Title'] = df_cleaned['Name'].str.extract(pat=r",\s*(Mr|Mrs|Miss)\b")
df_cleaned['Title'].unique()
```

### Group Analysis for Gaining Insights

```python
survived_rate = df_cleaned.groupby('Sex')['Survived'].mean()
print(survived_rate)
```

```python
class_group = df_cleaned.groupby('Pclass')['Survived'].mean()
print(class_group)
```

```python
rate = df_cleaned.groupby(['Sex', 'Pclass'])['Survived'].mean()
rate
```

### Binning Numerical Data


```python
def age_groups(age):
    if age < 18:
        return 'Child'
    elif 18 < age < 65:
        return 'Adult'
    else:
        return 'Senior'
    
df_cleaned['AgeGroup'] = df_cleaned['Age'].apply(age_groups)
df_cleaned
```
